This notebook details the procedure used to map the reads from the infection experiments to the database of short fragments, used for the mutations.

### Create the reference database

The original xls table was exported as a plain text table. The format of the table is:

| Running No | Target gene| gene in LT2 | description | Primer 1 | Primer 2|
|------|

For the unique id of each gene, the first column Running No, will be used

```python
#Import file, and create a fasta file for Primer1 and Primer 2
#The table is GeneSeqsTable.txt
!mkdir GeneSeqs_DB

#I will store the info of this table in a hash
#This will be used to later add the information for the outputs

table_dict_info = defaultdict(list)

with open("GeneSeqTable.txt") as fp:
    for line in fp:
        line = line.rstrip()
        if line.strip():
            running_no, target_gene, gene_LT2, desc, primer1, primer2 = line.split("\t")
            
            #I will use the running no as the ID to store in the dictionary. I need to check also
            #that is not present
            
            if running_no in table_dict_info.keys():
                print line 
                print "Already there"
                break
            else:
                table_dict_info[running_no] = [target_gene, gene_LT2, desc, primer1, primer2]
            
```

```python
#Using the dictionary in memory, now create the output for primer 1 and primer 2

#Open the files
primer1_file = open("GeneSeqs_DB/primer1.fna", 'w')
primer2_file = open("GeneSeqs_DB/primer2.fna", 'w')

for entry in table_dict_info:
    primer1_seq = table_dict_info[entry][3]
    primer2_seq = table_dict_info[entry][4]
    
    primer1_file.write(">" + entry + "\n")
    primer2_file.write(">" + entry + "\n")
    
    primer1_file.write(primer1_seq + "\n")
    primer2_file.write(primer2_seq + "\n")

primer1_file.close()
primer2_file.close()
    
    
```

```python
#Index database using Bowtie2

!bowtie2-build GeneSeqs_DB/primer1.fna GeneSeqs_DB/P1
!bowtie2-build GeneSeqs_DB/primer2.fna GeneSeqs_DB/P2
```

For bowtie2, I'm using the following parameters

    -k 1 Allows only one match
    -p 2 Number of threads. I'm running this on my laptop, so just using 2 threads
    --very-sensitive-local Uses local alignment
    -x Reference database
    -1 and -2 R1 and R2 of the reads
    
    
    

```python
#Create folder
!mkdir mapped_reads


#Map the In reads to P1
!bowtie2 -k 1 -p 2 --very-sensitive-local -x GeneSeqs_DB/P1 -1 seqs/Salmonella-input_S3_L001_R1_001.fastq.gz -2 seqs/Salmonella-input_S3_L001_R2_001.fastq.gz -S mapped_reads/input_P1.sam
#Map the In reads to P2
!bowtie2 -k 1 -p 2 --very-sensitive-local -x GeneSeqs_DB/P2 -1 seqs/Salmonella-input_S3_L001_R1_001.fastq.gz -2 seqs/Salmonella-input_S3_L001_R2_001.fastq.gz -S mapped_reads/input_P2.sam

#Map the Out reads to P1
!bowtie2 -k 1 -p 2 --very-sensitive-local -x GeneSeqs_DB/P1 -1 seqs/Salmonella-pi_S4_L001_R1_001.fastq.gz -2 seqs/Salmonella-pi_S4_L001_R2_001.fastq.gz -S mapped_reads/pi_P1.sam
#Map the Out reads to P2
!bowtie2 -k 1 -p 2 --very-sensitive-local -x GeneSeqs_DB/P2 -1 seqs/Salmonella-pi_S4_L001_R1_001.fastq.gz -2 seqs/Salmonella-pi_S4_L001_R2_001.fastq.gz -S mapped_reads/pi_P2.sam

```

## Looking at the output files (SAM format)

First I will create the bam files (a compressed version of SAM). This will help me to save hard drive space. The SAM files created by Bowtie2, have ~2GB each, while the BAM version is ~500Mb. 

```python
#Create bam files
!samtools view -b -o mapped_reads/input_P1.bam mapped_reads/input_P1.sam
!samtools view -b -o mapped_reads/input_P2.bam mapped_reads/input_P2.sam
!samtools view -b -o mapped_reads/pi_P1.bam mapped_reads/pi_P1.sam
!samtools view -b -o mapped_reads/pi_P2.bam mapped_reads/pi_P2.sam
```

```python
#Delete sam files to free space
!rm mapped_reads/input_P1.sam
!rm mapped_reads/input_P2.sam
!rm mapped_reads/pi_P1.sam
!rm mapped_reads/pi_P2.sam
```

```python
#Create bam files only with the reads that had matches, sort them and index.
!samtools view -b -F 4 -o mapped_reads/aln.input_P1.bam mapped_reads/input_P1.bam
!samtools sort -T temp -o mapped_reads/sorted.aln.input_P1.bam mapped_reads/aln.input_P1.bam
!samtools index mapped_reads/sorted.aln.input_P1.bam

!samtools view -b -F 4 -o mapped_reads/aln.input_P2.bam mapped_reads/input_P2.bam
!samtools sort -T temp -o mapped_reads/sorted.aln.input_P2.bam mapped_reads/aln.input_P2.bam
!samtools index mapped_reads/sorted.aln.input_P2.bam

!samtools view -b -F 4 -o mapped_reads/aln.pi_P1.bam mapped_reads/pi_P1.bam
!samtools sort -T temp -o mapped_reads/sorted.aln.pi_P1.bam mapped_reads/aln.pi_P1.bam
!samtools index mapped_reads/sorted.aln.pi_P1.bam

!samtools view -b -F 4 -o mapped_reads/aln.pi_P2.bam mapped_reads/pi_P2.bam
!samtools sort -T temp -o mapped_reads/sorted.aln.pi_P2.bam mapped_reads/aln.pi_P2.bam
!samtools index mapped_reads/sorted.aln.pi_P2.bam

```

```python
#Get the number of reads that mapped to each reference, using samtools
!samtools idxstats mapped_reads/sorted.aln.input_P1.bam > mapped_reads/stats.input_P1
!samtools idxstats mapped_reads/sorted.aln.input_P2.bam > mapped_reads/stats.input_P2
!samtools idxstats mapped_reads/sorted.aln.pi_P1.bam > mapped_reads/stats.pi_P1
!samtools idxstats mapped_reads/sorted.aln.pi_P2.bam > mapped_reads/stats.pi_P2
```

```python
#Read each file, and store the values in a dictionary
count_input_P1 = defaultdict()
count_input_P2 = defaultdict()
count_pi_P1 = defaultdict()
count_pi_P2 = defaultdict()

with open("mapped_reads/stats.input_P1") as fp:
    for line in fp:
        line = line.rstrip()
        gene_id, length, count, not_mapped = line.split("\t")
        count_input_P1[gene_id] = int(count)

with open("mapped_reads/stats.input_P2") as fp:
    for line in fp:
        line = line.rstrip()
        gene_id, length, count, not_mapped = line.split("\t")
        count_input_P2[gene_id] = int(count)
        
with open("mapped_reads/stats.pi_P1") as fp:
    for line in fp:
        line = line.rstrip()
        gene_id, length, count, not_mapped = line.split("\t")
        count_pi_P1[gene_id] = int(count)
        
with open("mapped_reads/stats.pi_P2") as fp:
    for line in fp:
        line = line.rstrip()
        gene_id, length, count, not_mapped = line.split("\t")
        count_pi_P2[gene_id] = int(count)
```

```python
#Make a big table output, using the original table as reference
#I will read that table again, just to keep the original order

table_with_counts = open("GeneTable_WithCounts.txt", 'w')
table_with_ra = open("GeneTable_WithRelativeAbundance.txt", 'w')

total_input_P1 = sum(count_input_P1.values())
total_input_P2 = sum(count_input_P2.values())
total_pi_P1 = sum(count_pi_P1.values())
total_pi_P2 = sum(count_pi_P2.values())

with open("GeneSeqTable.txt") as fp:
    for line in fp:
        line = line.rstrip()
        if line.strip():
            running_no, target_gene, gene_LT2, desc, primer1, primer2 = line.split("\t")
            
            #I will use the running no as the ID to store in the dictionary. I need to check also
            #that is not present
            
            #Get the values
            if running_no in count_input_P1:
                input_p1 = count_input_P1[running_no]
                ra_input_p1 = input_p1 / float(total_input_P1) * 100
            else:
                input_p1 = 0
                ra_input_p1 = 0
            
            if running_no in count_input_P2:
                input_p2 = count_input_P2[running_no]
                ra_input_p2 = input_p2 / float(total_input_P2) * 100
            else:
                input_p2 = 0
                ra_input_p2 = 0
                
            if running_no in count_pi_P1:
                pi_p1 = count_pi_P1[running_no]
                ra_pi_p1 = pi_p1 / float(total_pi_P1) * 100
            else:
                pi_p1 = 0
                ra_pi_p1 = 0
            
            if running_no in count_pi_P2:
                pi_p2 = count_pi_P2[running_no]
                ra_pi_p2 = pi_p2 / float(total_pi_P2) * 100
            else:
                pi_p2 = 0
                ra_pi_p2 = 0
                
            #Print everything to the new file
            out_values = [input_p1, input_p2, pi_p1, pi_p2]
            #out_relative_abundace = [ra_input_p1, ra_input_p2, ra_pi_p1, ra_pi_p2]
            
            table_with_counts.write("\t".join(map(str, out_values)) + "\t" + line + "\n")
            
            table_with_ra.write("%.3f\t%.3f\t%.3f\t%.3f\t" % (ra_input_p1, ra_input_p2, ra_pi_p1, ra_pi_p2))
            table_with_ra.write(line + "\n")
            
table_with_counts.close()
table_with_ra.close()
```
